```python
#Calculate the up and down
for i in range(len(df_twii['Close'])):
    if i+1 < len(df_twii['Close']):
        df_twii.loc[df_twii.index[i+1],'CloseP'] =\
        (df_twii.loc[df_twii.index[i+1],'Close']-df_twii.loc[df_twii.index[i],'Close'])/df_twii.loc[df_twii.index[i],'Close']
        
for i in range(len(df_0050['Close'])):
    if i+1 < len(df_0050['Close']):
        df_0050.loc[df_0050.index[i+1],'CloseP'] =\
        (df_0050.loc[df_0050.index[i+1],'Close']-df_0050.loc[df_0050.index[i],'Close'])/df_0050.loc[df_0050.index[i],'Close']
        
for i in range(len(df_tsmc['Close'])):
    if i+1 < len(df_tsmc['Close']):
        df_tsmc.loc[df_tsmc.index[i+1],'CloseP'] =\
        (df_tsmc.loc[df_tsmc.index[i+1],'Close']-df_tsmc.loc[df_tsmc.index[i],'Close'])/df_tsmc.loc[df_tsmc.index[i],'Close']
```

```python
#View the up and down
df_all_p = pd.DataFrame({'twii_p':df_twii['CloseP'],
                        '0050_p':df_0050['CloseP'],
                        'tsmc_p':df_tsmc['CloseP']})

plt.figure()
ax2 = df_all_p.plot(figsize=(10,4),grid=False, alpha = 0.5)
plt.show()
```

```python
#Add the sign
for p in range(len(df_all_p['twii_p'])):
    if df_all_p.loc[df_all_p.index[p],'0050_p'] >= 0:
        df_all_p.loc[df_all_p.index[p],'0050_s'] = '+'
    else:
        df_all_p.loc[df_all_p.index[p],'0050_s'] = '-'
    
    if df_all_p.loc[df_all_p.index[p],'tsmc_p'] >= 0:
        df_all_p.loc[df_all_p.index[p],'tsmc_s'] = '+'
    else:
        df_all_p.loc[df_all_p.index[p],'tsmc_s'] = '-'
    
    if df_all_p.loc[df_all_p.index[p],'twii_p'] >= 0:
        df_all_p.loc[df_all_p.index[p],'twii_s'] = '+'
    else:
        df_all_p.loc[df_all_p.index[p],'twii_s'] = '-'
```

```python
df_all_p.index = pd.to_datetime(df_all_p.index)
```

```python
#Slice the data with the date have three price
df_all_s = df_all_p.ix['20080103':'20171025']
```

```python
#clean the null data
df_all_s = df_all_s.dropna()
```

```python
df_all_s.head()
```

```python
#Difference times between 0050 and twii
diff = 0
reverse = 0
grow = 0
down = 0
for i in range(len(df_all_s)):
    if (df_all_s.ix[i,'0050_s'] == df_all_s.ix[i,'twii_s'])==False:
        diff += 1
        if i+1 < len(df_all_s):
            if df_all_s.ix[i+1,'0050_s'] == df_all_s.ix[i+1,'twii_s']:
                reverse += 1
                if df_all_s.ix[i+1,'0050_s'] == '+':
                    grow +=1
                else:
                    down +=1

print('Total data Qty:', len(df_all_s))
print('The different time:',diff)
print('Reverse to same:',reverse)
print('Reverse rate:',reverse/diff)
print('Reverse to grow:', grow, ',',grow/reverse)
print('Reverse to down:', down,',',1-grow/reverse)
```

```python
#Difference times between 0050 and twii and tsmc
diff = 0
reverse = 0
grow = 0
down = 0
for i in range(len(df_all_s)):
    if (df_all_s.ix[i,'0050_s'] == df_all_s.ix[i,'twii_s'] == df_all_s.ix[i,'tsmc_s'])==False:
        diff += 1
        if i+1 < len(df_all_s):
            if df_all_s.ix[i+1,'0050_s'] == df_all_s.ix[i+1,'twii_s'] == df_all_s.ix[i+1,'tsmc_s']:
                reverse += 1
                if df_all_s.ix[i+1,'0050_s'] == '+':
                    grow +=1
                else:
                    down +=1

print('Total data Qty:', len(df_all_s))
print('The different time:',diff)
print('Reverse to same:',reverse)
print('Reverse percentage:',reverse/diff)
print('Reverse to grow:', grow, ',',grow/reverse)
print('Reverse to down:', down,',',1-grow/reverse)
```
